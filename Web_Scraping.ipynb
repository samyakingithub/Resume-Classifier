```python
import pandas as pd
```

```python
str = "https://in.indeed.com"
df = pd.DataFrame(columns=["Job Title","Company Job Title","Company Link","Company Name","Company Location"])
def func(job_input):
    driver.get('https://in.indeed.com/jobs?q='+job_input+'&l=India')
    soup = bs.BeautifulSoup(driver.page_source, 'html.parser')
    for ele in soup.find_all(class_ = "css-1m4cuuf e37uo190"):
        ele = ele.parent()
        #     print(ele[6])

        job = ele[0].find(class_ = "jcs-JobTitle css-jspxzf eu4oa1w0")
        s1 = str+job["href"]
        #         print(f'Job Title: {job.text}')    #Job Title
        #         print(f'Company link: {s1}') #Company Link

        cn = ele[6].find(class_ ="turnstileLink companyOverviewLink")
        if(cn!=None):
            cn = cn.text
        #             print(f'Company Name: {cn.text}')
        lc = ele[4].find(class_ = "companyLocation")
        if(lc!=None):
            lc = lc.text
        #             print(f'Company Location: {lc.text}')
        #         print("\n") 
        df.loc[len(df)] = [job_input,job.text,s1,cn,lc]
```

```python
categories = ["Data Science","HR","Advocate","Arts","Web Designing","Mechanical Engineer","Sales","Health and fitness","Civil Engineer","Java Developer","Business Analyst","SAP Developer","Automation Testing","Electrical Engineering","Operations Manager","Python Developer","DevOps Engineer","Network Security Engineer","PMO","Database","Hadoop","ETL Developer","DotNet Developer","Blockchain","Testing"]
for i in categories:
    print(i)
    func(i)
```
